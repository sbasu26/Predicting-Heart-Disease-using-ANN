```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X_train = pd.read_csv("train_values.csv")
Y_train = pd.read_csv("train_labels.csv")
```

```python
X_train.head()
```

```python
Y_train.head()
```

```python
X_train = X_train.iloc[:,1:15].values
```

```python
Y_train = Y_train.iloc[:,1].values
```

```python
X_test = pd.read_csv("test_values.csv")
```

```python
X_test.head()
```

```python
X_test = X_test.iloc[:,1:15].values
X_test.shape
```

```python
## Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
X_train[:, 1] = labelencoder_X_1.fit_transform(X_train[:, 1])
labelencoder_X_2 = LabelEncoder()
X_test[:, 1] = labelencoder_X_2.fit_transform(X_test[:, 1])
```

```python
X_train.astype(float)
```

```python
X_test.astype(float)
```

```python
X_test.shape
```

```python
onehotencoder = OneHotEncoder(categorical_features = [1])
X_train = onehotencoder.fit_transform(X_train).toarray()
X_train = X_train[:, 1:]

onehotencoder2 = OneHotEncoder(categorical_features = [1])
X_test = onehotencoder2.fit_transform(X_test).toarray()
X_test = X_test[:, 1:]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_test.shape
```

```python
#import tensorflow
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Initialising the ANN
classifier = Sequential()
```

```python
#Adding the input layer and the first hidden layer
classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu', input_dim = 14))
#Adding the second hidden layer
classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu'))
#Adding the third hidden layer
classifier.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu'))
#Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
#Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
#Fitting the ANN to the training set
classifier.fit(X_train, Y_train, batch_size = 30, epochs = 200)
```

```python
#Predicting the test set results
y_pred = classifier.predict(X_test)
```

```python
y_pred = (y_pred > 0.8)
```

```python
y_pred = y_pred.astype(float)
```

```python
final_data = pd.DataFrame(data=y_pred[0:,1:],    # values
              index=y_pred[0:,0],    # 1st column as index
             columns=y_pred[0,1:]) 
```

```python
final_data.to_csv('final_submission96.csv', sep='\t', encoding='utf-8')
```
